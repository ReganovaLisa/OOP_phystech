Напишите функцию, которая для двух аргументов - строки РНА и строки-триплета будет возвращать количество вхождений триплета в РНА. (например для РНА "bccbcbcb" количество вхождений триплета "cbc" равно 2ум) **(2 балла)**

````python
def num_triplets(rna, triplet):
    num = 0
    for i in range(0,len(rna) - 3):
        if (rna[i] + rna[i+1] + rna[i+2]) == triplet:
            num +=1
    return(num)
#print(num_triplets('bccbcbcb', 'cbc'))
````
